```python
#importing required libraries

import numpy as np 
import pandas as pd 

import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df = pd.read_csv("Shark Tank India.csv")
df.head(3)
```

## Number of episodes and pitches in each season

```python
# Group the data by "Season Number" and get the counts
season_counts = df.groupby('Season Number').agg({'Episode Number': 'nunique', 'Pitch Number': 'count'})

# Renaming the columns for clarity
season_counts = season_counts.rename(columns={'Episode Number': 'Number of Episodes', 'Pitch Number': 'Number of Pitches'})

# Display the season counts with aligned headings
print(season_counts.to_string(justify='left'))
```

#### Season 1 had 37 episodes with 152 Startups that pitched to the sharks

#### Season 2 had 51 episodes with 168 Startups that pitched to the sharks

## Industry-wise Startups

```python
shark_tank_season1 = df.loc[df['Season Number']==1]
shark_tank_season2 = df.loc[(df['Season Number']==2)]
```

```python

import matplotlib.pyplot as plt

# Calculate the value counts of industries in season 1
industry_counts = shark_tank_season1['Industry'].value_counts().sort_values(ascending=True)

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.barh(industry_counts.index, industry_counts.values, color='steelblue')
plt.xlabel('Number of Startups')
plt.ylabel('Industry')
plt.title('Shark Tank India Season 1 - Industry-wise Startups')
plt.tight_layout()

# Show the graph
plt.show()
```

```python
# Calculate the value counts of industries in season 2
industry_counts2 = shark_tank_season2['Industry'].value_counts().sort_values(ascending=True)

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.barh(industry_counts2.index, industry_counts2.values, color='steelblue')
plt.xlabel('Number of Startups')
plt.ylabel('Industry')
plt.title('Shark Tank India Season 2 - Industry-wise Startups')
plt.tight_layout()

# Show the graph
plt.show()
```

## Top 40 Investments

```python
# Top 40 investments, as per total investment/deal amount (in lakhs)
print(df.groupby('Startup Name')['Total Deal Amount'].max().nlargest(40))


# Select the top 40 investments based on the total deal amount
top_investments = df.sort_values('Total Deal Amount', ascending=False).head(40)

# Generate an array of colors for each bar
colors = np.random.choice(list(plt.cm.colors.CSS4_COLORS.keys()), len(top_investments))

# Create a vertical bar graph with black background
plt.figure(figsize=(12, 8), facecolor='k')
ax = plt.gca()
ax.set_facecolor('k')
plt.bar(top_investments['Startup Name'], top_investments['Total Deal Amount'], color=colors)
plt.xlabel('Startup Name', color='white')
plt.ylabel('Total Deal Amount (in lakhs)', color='white')
plt.title('Top 40 Investments based on Total Deal Amount', color='white')
plt.xticks(rotation=90, color='white')
plt.yticks(color='white')
plt.tight_layout()

```

## Amount Invested by sharks

```python

# Season 1
Amount_s1 = [shark_tank_season1['Ashneer Investment Amount'].sum(), shark_tank_season1['Namita Investment Amount'].sum(), shark_tank_season1['Anupam Investment Amount'].sum(),
             shark_tank_season1['Vineeta Investment Amount'].sum(), shark_tank_season1['Aman Investment Amount'].sum(), shark_tank_season1['Peyush Investment Amount'].sum(),
             shark_tank_season1['Ghazal Investment Amount'].sum()]
name_s1 = ['Ashneer', 'Namita', 'Anupam', 'Vineeta', 'Aman', 'Peyush', 'Ghazal']

# Season 2
Amount_s2 = [shark_tank_season2['Namita Investment Amount'].sum(), shark_tank_season2['Anupam Investment Amount'].sum(), shark_tank_season2['Vineeta Investment Amount'].sum(),
             shark_tank_season2['Aman Investment Amount'].sum(), shark_tank_season2['Peyush Investment Amount'].sum(), shark_tank_season2['Amit Investment Amount'].sum()]
name_s2 = ['Namita', 'Anupam', 'Vineeta', 'Aman', 'Peyush', 'Amit']

# Set the style
plt.style.use('seaborn-darkgrid')

# Create subplots for both seasons
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot for Season 1
ax1.bar(name_s1, Amount_s1, color='steelblue')
ax1.set_xticklabels(name_s1, rotation=45, ha='right')
ax1.set_ylabel('Total Amount (in lakhs)', fontsize=12)
ax1.set_title('Total Investment Amount by Sharks - Season 1', fontsize=14)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
for index, d in enumerate(Amount_s1):
    ax1.text(x=index, y=d + 2, s=f"{round(d, 2)}", ha='center', fontsize=10)

# Plot for Season 2
ax2.bar(name_s2, Amount_s2, color='steelblue')
ax2.set_xticklabels(name_s2, rotation=45, ha='right')
ax2.set_ylabel('Total Amount (in lakhs)', fontsize=12)
ax2.set_title('Total Investment Amount by Sharks - Season 2', fontsize=14)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
for index, d in enumerate(Amount_s2):
    ax2.text(x=index, y=d + 2, s=f"{round(d, 2)}", ha='center', fontsize=10)

# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.4);

# Show the graph
plt.show();
```

## Trend of total investment amount by season

```python
total_investment_by_season = df.groupby('Season Number')['Total Deal Amount'].sum()
print("Total Investment Amount by Season:")
print(total_investment_by_season)

print("")
print("")

plt.figure(figsize=(8, 6))
plt.plot(total_investment_by_season.index, total_investment_by_season.values, marker='o')
plt.xlabel('Season Number')
plt.ylabel('Total Investment Amount (in lakhs)')
plt.title('Total Investment Amount by Season')
plt.xticks(total_investment_by_season.index)
plt.grid(True)
plt.show()
```

## Distribution of startups across industries (TOP 5)

```python
# Count the number of startups in each industry
startup_counts_by_industry = df['Industry'].value_counts()

# Select the top 5 industries
top_industries = startup_counts_by_industry.head(5)

# Create a pie chart to visualize the distribution of startups across industries
plt.figure(figsize=(8, 8))
plt.pie(top_industries, labels=top_industries.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Startups across Industries')
plt.axis('equal')
plt.show()
```
